```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('output.csv')
```

```python
# df = df.drop_duplicates(subset='Key')
```

```python
df.shape[0]
```

```python
sno = df['SNo']
size = df['Size']
shiftCount = df['ShiftCount']
comparisonCount = df['ComparisonCount']
```

```python
# Plotting ShiftCount
plt.figure(figsize=(8, 6))
plt.plot(size, shiftCount, color='orange')
plt.xlabel('size')
plt.ylabel('ShiftCount')
plt.title('ShiftCount')
plt.grid(True)
plt.show()

```

```python
# Create a figure
fig, ax1 = plt.subplots(figsize=(8, 6))

# Plot ShiftCount on the left y-axis
ax1.plot(size, shiftCount, color='orange')
ax1.set_xlabel('size')
ax1.set_ylabel('ShiftCount', color='orange')
ax1.tick_params(axis='y', labelcolor='orange')

# Create a secondary y-axis
ax2 = ax1.twinx()

# Plot ComparisonCount on the right y-axis
ax2.plot(size, comparisonCount, color='blue')
ax2.set_ylabel('ComparisonCount', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Set title
plt.title('Comparison of ShiftCount and ComparisonCount')

# Add grid
ax1.grid(True)

# Show the combined plot
plt.show()
```
