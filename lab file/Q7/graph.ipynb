```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('output.csv')
```

```python
# df = df.drop_duplicates(subset='Key')
```

```python
df.shape[0]
```

```python
sno = df['SNo']
size = df['Size']
shiftCount = df['ShiftCount']
comparisonCount = df['ComparisonCount']
```

```python
# Plotting ShiftCount
plt.figure(figsize=(8, 6))
plt.plot(size, shiftCount, color='orange')
plt.xlabel('size')
plt.ylabel('ShiftCount')
plt.title('ShiftCount')
plt.grid(True)
plt.show()

```

```python

# Plotting ComparisonCount
plt.figure(figsize=(8, 6))
plt.plot(size, comparisonCount, color='blue')
plt.xlabel('size')
plt.ylabel('ComparisonCount')
plt.title('ComparisonCount')
plt.grid(True)
plt.show()
```
